```python
# インポート
import sqlite3
import pandas as pd
import os
```

```python
# ファイル内容取得
csv_filepath = "wine.csv"
df = pd.read_csv(csv_filepath, delimiter=",")
```

```python
# データベースに接続
sqlite_dirpath = "db"
# sqlite_filename = "test_0.sqlite"
sqlite_filename = "test_0.db"
sqlite_filepath = os.path.join(sqlite_dirpath, sqlite_filename)

os.makedirs(sqlite_dirpath, exist_ok=True)  # フォルダの作成
conn = sqlite3.connect(sqlite_filepath)  # ファイルがなければ生成されるが，フォルダは生成されないので注意
```

```python
# テーブルを作成
table_name = "wine"

col1_name = "Wine"
col2_name = "Alcohol"
col3_name = "MalicAcid"

col1_type = "INTEGER"
col2_type = "FLOAT"
col3_type = "FLOAT"

cur = conn.cursor()  # カーソルの呼び出し

# テーブルの初期化
drop_sql = f"DROP TABLE IF EXISTS {table_name}"
create_sql = f"""CREATE TABLE {table_name}(
    id INTEGER PRIMARY KEY,
    {col1_name} {col1_type},
    {col2_name} {col2_type},
    {col3_name} {col3_type}
)"""

# SQLの実行
cur.execute(drop_sql)
cur.execute(create_sql)
# SQLの確定
conn.commit()
```

```python
# データ挿入
for index, row in df.iterrows():
    cur.execute(
        f"INSERT INTO {table_name} (id, {col1_name}, {col2_name}, {col3_name}) VALUES (?, ?, ?, ?)",
        (index, row['Wine'], row['Alcohol'], row['Malic.acid'])
    )
    conn.commit()
    
    if index == 2:
        break
```

```python
# 中身の確認
select_sql = f"SELECT id, {col1_name}, {col2_name}, {col3_name} FROM {table_name}"
cur.execute(select_sql)

wine_data = cur.fetchall()
wine_data
```
